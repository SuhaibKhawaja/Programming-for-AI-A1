```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
grid = np.column_stack([[x, y] for x in range(-15, 16, 1) for y in range(-15, 16, 1)])

plt.figure(figsize=(10, 10), facecolor="w")
plt.scatter(grid[0], grid[1], s=36, edgecolor="none")
# Set axis limits
plt.grid(True)
plt.axis("equal")
plt.xlim(-30,30)
plt.ylim(-30,30)
plt.title("First Grid")
```

```python
grid = np.column_stack([[x, y] for x in range(-15, 16, 1) for y in range(-15, 16, 1)])

for i in range(len(grid[0])):
    grid[0][i] += grid[1][i]
        
plt.figure(figsize=(10, 10), facecolor="w")
plt.scatter(grid[0], grid[1], s=36, edgecolor="none")
plt.grid(True)
plt.axis("equal")
plt.xlim(-30,30)
plt.ylim(-30,30)
plt.title("Shear Transformed Gird With Loops")
```

```python
grid = np.column_stack([[x, y] for x in range(-15, 16, 1) for y in range(-15, 16, 1)])
grid[0] = grid[0]+grid[1]

plt.figure(figsize=(10, 10), facecolor="w")
plt.scatter(grid[0], grid[1], s=36, edgecolor="none")
plt.grid(True)
plt.axis("equal")
plt.xlim(-30,30)
plt.ylim(-30,30)
plt.title("Shear Transformed Gird Without any Loops")
```
